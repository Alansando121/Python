```python
import numpy as np
import random

# generr id
def generar_id():
    return random.randint(100000, 999999)

#automatico
def agregar_automatico(lista):
    nombres = ["Juan", "María", "Pedro", "Luis", "Ana", "Carlos", "Laura", "Diego", "Sofía", "Miguel", "Alejandro", "Paula", "Lucía", "Daniel", "Fernanda", "Andrés", "Valentina", "Pablo", "Gabriela"]
    apellidos = ["García", "Rodríguez", "Martínez", "Hernández", "López", "González", "Pérez", "Sánchez", "Ramírez", "Torres", "Fernández", "Díaz", "Moreno", "Álvarez", "Vázquez", "Ruiz", "Gómez", "Gutiérrez", "Navarro", "Martín"]
    cargos = ["Operario", "Gerente", "Empleado"]

    if len(lista) == 0:
        lista.append({'ID': generar_id(), 'Nombre': random.choice(nombres), 'Apellido': random.choice(apellidos), 'Edad': random.randint(18, 60), 'Cargo': random.choice(cargos)})
    else:
        ultimo_id = lista[-1]['ID']
        lista.append({'ID': generar_id(), 'Nombre': random.choice(nombres), 'Apellido': random.choice(apellidos), 'Edad': random.randint(18, 60), 'Cargo': random.choice(cargos)})
    print("Trabajador agregado correctamente.")

# manual
def agregar_manual(lista):
    id_repetido = True
    while id_repetido:
        nuevo_id = int(input("Ingrese el ID de 6 dígitos del trabajador: "))
        if len(str(nuevo_id)) != 6:
            print("Error: El ID debe tener 6 dígitos.")
            continue
        id_repetido = False
        for trabajador in lista:
            if trabajador['ID'] == nuevo_id:
                print("Error: El ID ingresado ya existe. Por favor, ingrese un ID diferente.")
                id_repetido = True
                break

    nombre = input("Ingrese el nombre del trabajador: ")
    apellido = input("Ingrese el apellido del trabajador: ")
    while True:
        edad = int(input("Ingrese la edad del trabajador (mayor de 18 años): "))
        if edad > 18:
            break
        else:
            print("Error: La edad debe ser mayor de 18 años.")
    cargo = input("Ingrese el cargo del trabajador (Operativo/Gerente/Empleado): ")
    while cargo not in ["Operativo", "Gerente", "Empleado"]:
        print("Error: Cargo no valido.")
        cargo = input("Ingrese el cargo del trabajador (Operativo/Gerente/Empleado): ")

    lista.append({'ID': nuevo_id, 'Nombre': nombre, 'Apellido': apellido, 'Edad': edad, 'Cargo': cargo})
    print("Trabajador agregado correctamente.")

# imprimir lista
def imprimir_lista(lista):
    if len(lista) == 0:
        print("La lista esta vacia.")
    else:
        for trabajador in lista:
            print(trabajador)

# Buscar id
def buscar_por_id(lista, id_busqueda):
    for trabajador in lista:
        if trabajador['ID'] == id_busqueda:
            print("Trabajador encontrado:")
            print(trabajador)
            return
    print("No se encontro ningun trabajador con el ID proporcionado.")

# ordenar por id
def ordenar_lista(lista):
    if len(lista) == 0:
        print("La lista esta vacia.")
    else:
        lista.sort(key=lambda x: x['ID'])
        print("Lista ordenada correctamente.")

# eliminae
def eliminar_por_id(lista, id_eliminar):
    for trabajador in lista:
        if trabajador['ID'] == id_eliminar:
            lista.remove(trabajador)
            print("Trabajador eliminado correctamente.")
            return
    print("No se encontro ningun trabajador con el ID proporcionado.")

# Borrar todo
def borrar_lista(lista):
    if len(lista) == 0:
        print("La lista esta vacia.")
    else:
        confirmacion = input("¿Esta seguro de que desea borrar toda la lista? (s/n): ")
        if confirmacion.lower() == 's':
            lista.clear()
            print("Lista borrada correctamente.")
        else:
            print("Operacion cancelada.")


lista_trabajadores = []

# }
while True:
    print("\nMENU\n")
    print("1.- Agregar (automatico)")
    print("2.- Agregar (manual)")
    print("3.- Imprimir lista")
    print("4.- Buscar por ID")
    print("5.- Ordenar lista")
    print("6.- Eliminar por ID")
    print("7.- Borrar toda la lista")
    print("0.- SALIR")

    opcion = input("\nSeleccione una opcion: ")

    if opcion == '1':
        agregar_automatico(lista_trabajadores)
    elif opcion == '2':
        agregar_manual(lista_trabajadores)
    elif opcion == '3':
        imprimir_lista(lista_trabajadores)
    elif opcion == '4':
        id_busqueda = int(input("Ingrese el ID del trabajador que desea buscar: "))
        buscar_por_id(lista_trabajadores, id_busqueda)
    elif opcion == '5':
        if len(lista_trabajadores) == 0:
            print("La lista esta vacia.")
        else:
            ordenar_lista(lista_trabajadores)
    elif opcion == '6':
        id_eliminar = int(input("Ingrese el ID del trabajador que desea eliminar: "))
        eliminar_por_id(lista_trabajadores, id_eliminar)
    elif opcion == '7':
        borrar_lista(lista_trabajadores)
    elif opcion == '0':
        print("¡Hasta luego!")
        break
    else:
        print("Opción no valida. Por favor, seleccione una opcion valida.")
1
```
